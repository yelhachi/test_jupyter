```python
from matplotlib import pylab as plt
from ImageD11.columnfile import columnfile as cf
import numpy as np
import weighted_kde
```

```python

data = cf('../input/r117-sample/frc_R117_700_1_261N_z0.flt.new')
```

```python
plt.plot(data.tth,data.eta,'rx',alpha=0.7,label=r'$all \, peaks$')
austenite = data.copy()
austenite.filter(austenite.labels>-1)
plt.plot(austenite.tth,austenite.eta,'gx',alpha=0.7,label=r'austenite')
plt.xlabel(r'$2 \theta (\degree)$')
plt.ylabel(r'$\eta (\degree)$')
plt.legend(loc='best')
plt.title('Radial transformation')
```

```python
martensite = data.copy()
martensite.filter(martensite.labels==-1)
```

```python
pdf = weighted_kde.gaussian_kde(austenite.tth, weights=austenite.sum_intensity)
x = np.linspace(min(data.tth), max(data.tth), 200)
y = pdf(x)
plt.plot(x, y)
plt.title(' For bandwidth = %.3f'%pdf.factor)
```

```python
pdf.set_bandwidth(pdf.factor/20.)
y = pdf(x)
plt.plot(x, y)
plt.title(' For bandwidth = %.3f'%pdf.factor)
```

```python
pdf_martensite = weighted_kde.gaussian_kde(martensite.tth, weights=martensite.sum_intensity)
pdf_martensite.set_bandwidth(0.04*pdf_martensite.factor)
y_austenite = pdf(x)
y_martensite = pdf_martensite(x)
#Normalize by taking into account the fraction of each phase
I_total = np.sum(data.sum_intensity)
I_austenite = np.sum(austenite.sum_intensity)/float(I_total)
I_martensite = np.sum(martensite.sum_intensity)/float(I_total)
plt.plot(x, y_austenite*I_austenite,label=r'$Austenite$')
plt.plot(x, y_martensite*I_martensite,label=r'$Martensite$')
plt.legend()
```

```python
pdf = weighted_kde.gaussian_kde(austenite.tth_per_grain, weights=austenite.sum_intensity*austenite.Lorentz_per_grain)
pdf.set_bandwidth(pdf.factor/20.)
x = np.linspace(min(data.tth), max(data.tth), 200)
y = pdf(x)
plt.plot(x, y)
plt.title(' For bandwidth = %.3f'%pdf.factor)
```

```python
pdf = weighted_kde.gaussian_kde(martensite.tth, weights=martensite.sum_intensity*martensite.Lorentz)
pdf.set_bandwidth(pdf.factor/20.)
x = np.linspace(min(data.tth), max(data.tth), 200)
y = pdf(x)
plt.plot(x, y)
plt.title(' For bandwidth = %.3f'%pdf.factor)
```
